```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('/data/examples/pm25/PM25_train.csv')
df
```

```python
print(len(df['device_id'].unique()))
print(len(df['Date'].unique()))
print(len(df['Time'].unique()))
```

```python
df1 = df.groupby(['device_id','Date']).agg({'PM2.5':'mean','Temperature':'mean','Humidity':'mean'}).unstack('Date')
df1
```

```python
df1.columns.levels[0]
```

```python
df2 = df1.copy()
for key in df1.columns.levels[0]:
    for i in range(252):
        for j in range(30):
            if np.isnan(df1[key].iloc[i,j]):
                df2[key].iloc[i,j] = df1[key].iloc[i].mean()/2 + df1[key].iloc[:,j].mean()/2
df2
```

```python
print(df.isnull().any().any())
print(df1.isnull().any().any())
print(df2.isnull().any().any())
```

```python
PEROID = 14

df3 = []
for i in range(30-PEROID):
    features_n_label = []
    for key in df2.columns.levels[0]:
        features_n_label.append(df2[key].iloc[:,i:i+PEROID])
    features_n_label.append(df2['PM2.5'].iloc[:,i+PEROID])
    samples = pd.concat(features_n_label, axis=1)
    samples.columns = range(PEROID*3+1)
    df3.append(samples)
df3 = pd.concat(df3)
X, y = df3.iloc[:,:PEROID*3], df3[PEROID*3].reshape(-1,1)

features = []
for key in df2.columns.levels[0]:
    features.append(df2[key].iloc[:,30-PEROID:])
test = pd.concat(features, axis=1)
test.columns = range(PEROID*3)
```

```python
print(X.shape)
print(y.shape)
print(test.shape)
```

```python
LR_model = LinearRegression(normalize=True)
LR_model.fit(X, y)
print(mean_squared_error(LR_model.predict(X), y))
print(r2_score(LR_model.predict(X), y))
```

```python
La_model = Lasso(normalize=True, alpha = 0.001)
La_model.fit(X, y)
print(mean_squared_error(La_model.predict(X), y))
print(r2_score(La_model.predict(X), y))
```

```python
Ri_model = Lasso(normalize=True, alpha = 0.01)
Ri_model.fit(X, y)
print(mean_squared_error(Ri_model.predict(X), y))
print(r2_score(Ri_model.predict(X), y))
```

```python
from xgboost import XGBRegressor
xgb_model = XGBRegressor(
                            learning_rate = 0.5,
                            n_estimators = 50,
                            max_depth= 6,
                            min_child_weight = 1,
                            gamma = 0,
                            subsample = 0.8,
                            colsample_bytree = 0.8,
                            objective = 'reg:linear',
                            reg_lambda = 0.8,
                            reg_alpha = 0,
                            nthread = 4,
                            scale_pos_weight = 1,
                            seed=99
                         )
xgb_model.fit(X, y)
print(mean_squared_error(xgb_model.predict(X), y))
print(r2_score(xgb_model.predict(X), y))
```

```python
submission_format = pd.read_csv('/data/examples/pm25/submission.csv')
submission_format['device_id'] == df1.index
```

```python
submission_format['pred_pm25'] = LR_model.predict(test)
submission_format.to_csv('pm25.csv', index=False, header=True)
```
