```python
from PyPDF2 import PdfReader, PdfWriter
reader = PdfReader('131.pdf')
EyeCanSee = PdfReader('EyeCanSee.pdf').pages[0]
writer = PdfWriter()
for page in reader.pages:
    if EyeCanSee:
        EyeCanSee.merge_page(page)
        EyeCanSee.mediabox = page.mediabox
        writer.add_page(EyeCanSee)
        EyeCanSee = 0
    else:
        writer.add_page(page)
with open("131.EyeCanSee.pdf", "wb") as fp:
    writer.write(fp)
```

```python
import ocrmypdf
ocrmypdf.ocr('131.EyeCanSee.pdf', '131.OCR.pdf', force_ocr=True)
```
