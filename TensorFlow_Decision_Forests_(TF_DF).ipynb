```python
import tensorflow_decision_forests as tfdf
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix
```

```python
df = pd.DataFrame({'feature': np.random.rand(1000)})
df['label'] = '大'
df
```

```python
df.loc[df['feature'] < 0.1, 'label'] = '小'
df
```

```python
df['label'].describe()
```

```python
train_ds = tfdf.keras.pd_dataframe_to_tf_dataset(df[:900], label='label')
test_ds = tfdf.keras.pd_dataframe_to_tf_dataset(df[900:], label='label')
```

```python
model = tfdf.keras.RandomForestModel(num_trees=10, random_seed=101)
```

```python
model.compile(metrics=["accuracy"])
model.fit(train_ds)
```

```python
tfdf.model_plotter.plot_model_in_colab(model, tree_idx=9, max_depth=7)
```

```python
evaluation = model.evaluate(test_ds)
```

```python
pred = model.predict(test_ds)
pred
```

```python
matrix = pd.DataFrame({'labels': df[900:]['label']})
matrix['predictions'] = pred
matrix.loc[matrix['predictions']==0, :] = '大'
matrix.loc[matrix['predictions']==1, :] = '小'
matrix
```

```python
matrix
```

```python
confusion_matrix(matrix['labels'], matrix['predictions'])
```

```python
tn, fp, fn, tp = confusion_matrix(matrix['labels'], matrix['predictions']).ravel()
tn, fp, fn, tp
```
