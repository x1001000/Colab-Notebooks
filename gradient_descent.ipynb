```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
```

```python
sample_size = 10
X = np.random.randint(100, 1000, sample_size)
斜率, 截距 = 55, 66
Y = 斜率 * np.random.normal(1, 0.1, sample_size) * X + 截距 * np.random.normal(1, 0.1, sample_size)
```

```python
model = LinearRegression()
model.fit(X.reshape(-1, 1), Y.reshape(-1, 1))
plt.scatter(X, Y)
plt.plot(X, model.predict(X.reshape(-1, 1)), color='red')
```

```python
Axes3D = plt.axes(projection='3d')
w_, b_ = np.meshgrid(range(100), range(100), indexing='ij')
Loss = []
for w in range(100):
    for b in range(100):
        loss = 0
        for n in range(sample_size):
            loss += ( w * X[n] + b - Y[n] ) ** 2
        Loss.append(loss)
Axes3D.plot_wireframe(w_, b_, np.array(Loss).reshape(100, 100))
```

```python
w = 0
b = 0
learning_rate = 0.1
learning_rate_w = 0
learning_rate_b = 0
iteration = 1000000
W = [w]
B = [b]
```

```python
for _ in range(iteration):
    Lw = 0 # partial derivative of L with respect to w
    Lb = 0 # partial derivative of L with respect to b
    for n in range(sample_size):
        Lw += 2 * ( w * X[n] + b - Y[n] ) * X[n]
        Lb += 2 * ( w * X[n] + b - Y[n] ) * 1
    learning_rate_w += Lw ** 2
    learning_rate_b += Lb ** 2
    w -= learning_rate / learning_rate_w ** 0.5 * Lw
    b -= learning_rate / learning_rate_b ** 0.5 * Lb
    W.append(w)
    B.append(b)
```

```python
plt.plot(W, B)
```
